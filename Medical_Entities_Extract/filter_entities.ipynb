```python
import pandas as pd
```

```python
import os
import json
from tqdm import tqdm
```

```python
with open('temp_dygie_output.json','r', encoding='utf-8') as f:
    data=json.load(f)
File_Path=[]
filtered_entities=[]
```

```python
for k,v in tqdm(data.items(),desc="Processing"):
    File_Path.append(k)
    if(len(v['entities'])==0):
       print(k)
       filtered_entities.append('No_Entity')
       continue
    entities=[]
    for k1,v1 in v['entities'].items():
        if ('DA' in v1['label']):
            continue
        entities.append(v1['tokens'])
    entities_toString=' '.join(list(set(entities)))
    filtered_entities.append(entities_toString)
```

```python
print(len(filtered_entities)==len(File_Path))
```

```python
df=pd.DataFrame()
```

```python
df['Path']=File_Path
df['Entities']=filtered_entities
```

```python
print(df)
```

```python
origin_df=pd.read_csv('mimic_report.csv',sep=',')
merge_df=pd.merge(left=origin_df,right=df,on="Path",how="inner")
merge_df.to_csv('final_training.csv',index=False)
print(len(merge_df)==len(origin_df))
```
